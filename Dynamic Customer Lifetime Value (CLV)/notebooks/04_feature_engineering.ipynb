# Notebook 04: Feature Engineering

## Objective
In this notebook, we:
- Aggregate transaction data at the customer level  
- Create behavioral, recency, frequency, and trend-based features  
- Construct the modeling dataset  
- Define the target variable: future 12-month customer lifetime value (CLV)

> Note: No ML modeling is performed here.

## Problem Context

Feature engineering transforms raw transactional logs into structured behavioral indicators.  
These features capture purchasing patterns, engagement levels, and trends that help predict future customer value.
