# Notebook 02: Data Cleaning

## Objective
In this notebook, we:
- Handle missing values  
- Correct data types (dates, numeric fields)  
- Remove invalid or impossible records  
- Address duplicates  
- Prepare clean datasets for analysis  

> Note: No feature engineering is performed here.

## Problem Context

Raw business data is rarely ready for modeling. Cleaning ensures:

- Data consistency  
- Reduced noise  
- Reliable downstream analysis  

This step improves model performance and trustworthiness.
