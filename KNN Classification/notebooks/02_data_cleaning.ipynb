```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("Libraries imported successfully.")
```

```python
df=pd.read_csv("../data/heart.csv")
print("Data loaded successfully.")
```

```python
df_missing = df.isnull().sum()
df_percent_missing = (df_missing / len(df)) * 100
missing_data=pd.DataFrame({
    "Columns":df_missing.index,
    "Missing Values":df_missing.values,
    "Percentage Missing":df_percent_missing.values
    }
)
print("Missing data calculated successfully.")
print(missing_data)
```

```python
df_duplicates=df.duplicated().sum()
print("Duplicate rows counted successfully.")
print(f"Data shape before removing duplicates: rows={df.shape[0]}, columns={df.shape[1]}")
print(f"Number of duplicate rows: {df_duplicates}")
if df_duplicates>0:
    print(df[df.duplicated()])
    df=df.drop_duplicates()
    print("Duplicate rows removed successfully.")
print(f"Data shape after removing duplicates: rows={df.shape[0]}, columns={df.shape[1]}")
```

```python
df.to_csv("../data/processed/cleaned_heart.csv", index=False)
print("Cleaned data saved successfully.")
```
